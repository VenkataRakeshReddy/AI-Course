```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import matplotlib.image as mpimg
```

```python
sns.set(style="darkgrid")
```

```python
#sns.get_dataset_names()
sns.get_dataset_names()
```

```python
iris=sns.load_dataset('iris')
```

```python
iris.head(10)
```

```python
#relplot with the orgument x,y,data
sns.regplot(x='sepal_length',y = 'sepal_width',data=iris)
```

```python
#relplot with the orugument x,y,data,hue
sns.relplot(x='sepal_length',y='sepal_width',data=iris,hue="petal_length")
```

```python
sns.scatterplot(x='sepal_length',y='sepal_width',data=iris,hue="petal_width")
```

```python
sns.relplot(x='petal_length',y='petal_width',data=iris,col="species",hue="petal_length",style="petal_width",kind="scatter")
```

```python
sns.relplot(x='sepal_length',y='sepal_width',data=iris,kind='line',ci=False)
```

```python
sns.relplot(x='sepal_length',y='sepal_width',data=iris,kind='line',ci=True)
```

```python
sns.relplot(x='sepal_length',y='sepal_width',data=iris,kind='line',hue='species')
```

```python
sns.regplot(x='sepal_length',y='sepal_width',data=iris)
```
